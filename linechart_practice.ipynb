```python
import matplotlib.pyplot as plt
```

```python
savings = [0, 100, 150, 200, 300, 500, 300, 350, 500, 600, 750, 1000]
months = ['jan', 'feb', 'march', 'april','may', 'june', 'july', 'august', 'sept', 'oct', 'nov', 'dec']
```

```python
plt.plot(months, savings)
plt.xlabel('Month')
plt.ylabel('Savings in Dollars')
plt.title('Savings Account Results')
plt.show()
```
